 Intern Data Validation System
Problem Statement:-
Digital Dreams HR team receives intern registration details. Create a Python program to validate intern eligibility.


```python
# Intern Data Validation System
# This program checks whether an intern is eligible for internship

# Step 1: Take input from the user
name = input("Enter Intern Name: ")
age = int(input("Enter Age: "))
email = input("Enter Email ID: ")
contact = input("Enter Contact Number: ")
percentage = float(input("Enter Graduation Percentage: "))

# Step 2: Check eligibility conditions
if age < 18:
    print("Intern Rejected: Age must be 18 or above")

elif percentage < 60:

    print("Intern Rejected: Graduation percentage must be 60% or above")

elif len(contact) != 10:
    print("Intern Rejected: Contact number must have exactly 10 digits")

else:
    print("Intern Eligible for Internship")
```
