Employee Salary Calculation System

Problem Statement:-
Digital Dreams requires a basic payroll calculation tool.

```python
# Employee Salary Calculation System
# This program calculates the net salary of an employee

# Step 1: Take input from the user
employee_name = input("Enter Employee Name: ")
employee_id = input("Enter Employee ID: ")
basic_salary = float(input("Enter Basic Salary: "))

# Step 2: Calculate salary components
hra = 0.20 * basic_salary     # 20% of Basic Salary
da = 0.10 * basic_salary      # 10% of Basic Salary
pf = 0.12 * basic_salary      # 12% of Basic Salary

# Step 3: Calculate Net Salary
net_salary = basic_salary + hra + da - pf

# Step 4: Display Salary Summary
print("\n----- Salary Summary -----")
print("Employee Name   :", employee_name)
print("Employee ID     :", employee_id)
print("Basic Salary    :", basic_salary)
print("HRA (20%)       :", hra)
print("DA (10%)        :", da)
print("PF (12%)        :", pf)
print("Net Salary      :", net_salary)
```
