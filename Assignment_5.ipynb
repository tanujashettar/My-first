 Employee Bonus Evaluation System


Problem Statement
Digital Dreams calculates annual bonus based on employee performance.

```python
name = input("Enter Employee Name: ")
salary = float(input("Enter Salary: "))
rating = int(input("Enter Performance Rating (1-5): "))

if rating == 5:
    bonus = 0.20 * salary
elif rating == 4:
    bonus = 0.15 * salary
elif rating == 3:
    bonus = 0.10 * salary
else:
    bonus = 0

print("Name:", name)
print("Rating:", rating)
print("Bonus:", bonus)
print("Final Salary:", salary + bonus)
```
