Course Fee Billing System

Problem Statement
Digital Dreams offers multiple training programs. Build a fee calculator.

```python
course = input("Enter Course Name: ")
# Standardize input for comparison to make it more robust (e.g., 'AI&ML' vs 'AI & ML')
normalized_course = course.lower().replace(" ", "").replace("&", "")

fee = None # Initialize fee to None

if normalized_course == "pythonprogramming":
    fee = 5000
elif normalized_course == "dataanalytics":
    fee = 8000
elif normalized_course == "aiml":
    fee = 12000

if fee is None: # If fee is still None, it means the course was not recognized
    print("Invalid Course")
    # In a Colab cell, 'exit()' might not immediately stop subsequent lines.
    # By putting calculations in an else block, we prevent NameError.
    exit() # Re-activate exit to prevent further execution for invalid courses
else:
    student_discount = 0.10 * fee
    early_discount = 0.05 * fee
    total_discount = student_discount + early_discount

    final_amount = fee - total_discount

    print("Course Name:", course)
    print("Original Fee: ₹", fee)
    print("Total Discount: ₹", total_discount)
    print("Final Payable Amount: ₹", final_amount)
```
